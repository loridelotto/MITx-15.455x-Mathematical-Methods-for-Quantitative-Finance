# The random Walk

## Time series models

A stochastic process is a time-dependent random variable:
- Continuous time $S(t)$
- Discrete time $S_1, S_2, S_3, \cdots , S_t$

For now we will consider only discrete time, where the variable is sampled at uniform intervals.

Discrete time processes can be constructed by adding increments, for example: $S_t = S_{t-1} + x_t = x_1 + x_2 + \cdots + x_t$

Increments can be recovered by differencing: $x_t = S_t - S_{t-1}$

Time series are used to model processes that evolve or are observed discretely in time, the model is "solved", when we can describe or forecast the distribution of not yet observed values. The most basic model is the random walk.

## Random walk model

The random walk is the simplest of all time-series models 
$$S_T = z_1 + z_2 + \cdots + z_T$$
Each increment is a random IID variable, which doesn't depend on past history. The evolution in time is uniform (stationarity), this means that if we look at the values between any two points in time, the properties depend only on the differences in the point in time, not on some absolute point in time.

**Properies of the random variables**

We can often express results in terms of "standard" random variables:
- mean = 0 $\quad \Rightarrow \quad E[z_t] = 0$
- variance = 1  $\quad \Rightarrow \quad E[z_t^2] = 1$
- correlation = 0  for different times $\quad \Rightarrow \quad E[z_t z_{t'}] = 0 \ \text{if} \  t\neq t'$

Examples:
- Coin toss (+1 or -1 with equal probability) $\quad z_t = \pm 1, \quad \quad p=1/2$
- Gaussian random variable $\quad z_t \thicksim \mathcal{N}(0,1), \quad \quad p(z_t) = \frac{1}{\sqrt{2\pi}} e^{-z_t^2/2}$

Now consider a **sum** of these identical random variables. What is the probability of the sum? Instead of computing the convolution of the probability distributions, we can just calculate the mean and the variance of the sum:

$$E[S_T] = E[z_1] + E[z_2] + \cdots + E[z_T] = 0 $$
$$\text{Var}(S_T) = E[S_T^2] = E[(z_1 + z_2 + \cdots + z_T)^2] = \sum^{T}_{t=1}E[z_t^2] + 2 \sum_{t<t'}E[z_t z_t'] = T$$

The random walk has variance that grows linearly with T, this means that its standard deviation grows as square root of T.


## Generalized random walk model

We can generalize this model by linearly rescaling our random variable:
$$r_t = \sigma z_t + \mu$$
The parameters $\sigma$ and $\mu$ measure the risk and return.

Application: **asset price returns**

Consider a series of regularly observed stock prices, whose logarithmic returns are defined by $$r_t = log(P_t/P_{t-1})$$
Then the price at time T will be $$P_T = P_o^{r_1 + r_2 + \cdots + r_T}$$

Since the expectation operator is linear, we can compute the expected return, variance, and covariance immediately from our previous results. 
$$E[r_t] = \mu$$
$$E[(r_t - \mu)^2] = \sigma^2$$
$$E[(r_t - \mu)(r_{t'}-\mu)] = 0 \ \text{if} \ t \neq t'$$

Now let's consider the stochastic process generated by the sum 
$$X_T = r_1 + r_2 + \cdots + r_T = \sum^{T}_{t=q} r_t$$
and let's compute the mean and variance of this sum:
- Mean: $E[X_t] = E[r_1] + E[r_2] + \cdots + E[r_T] = T\mu$
- Variance: $\text{Var}(X_T) = E[(X_T - T\mu)^2] = E[((r_1 - \mu)+(r_2 - \mu) + \cdots + (r_T - \mu))^2] = T\sigma^2$

## Linear time series models

To capture causality and correlation across time, more complew models can be built by combining elements of the random walk.

### Moving Average MA(1)

The variables are not IID anymore: $$r_t = \mu +\sigma z_t + \phi z_{t-1}$$

### Garch Model

The distribution may itself be time-varying: $$r_t = \mu +\sigma z_t = \mu + \epsilon_t, \quad \quad \epsilon_t \thicksim \mathcal{N}(0, \sigma_t^2)$$

### Autoregressive model of order p AR(p)

$$R_t = c_0 + c_1R_{t-1} + \cdots + c_p R_{t-p} + \sigma z_t, \quad \quad z_t\thicksim \text{IID}(0,1) $$

### Arma(p,q)

$$R_t = c_0 + c_1 R_{t-1} + \cdots + c_p R_{t-p} + \sigma z_t + \phi_1 z_{t-1} + \cdots + \phi_q z_{t-q}$$

The distribution of future outcomes can be determined using linearity, recusion and stationarity. 

**Stationarity** says that the joint distribution of all of its values is invariant under time translation $ t \rightarrow t+s $
It is **wakly stationary** if the first and second moments are invariant.

Example: AR(1)

$$R_t = c_0 + c_1 R_{t-1} + \sigma z_t$$
