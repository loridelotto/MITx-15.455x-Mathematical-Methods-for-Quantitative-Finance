# Model Estimation and Identification

## 1. Statistics vs Estimator


### Statistic
A **statistic** is a function of observed data.

Examples (given a time series of returns $r_t$):
$$
\hat{\mu} = \frac{1}{T} \sum_{t=1}^{T} r_t
$$
$$
\hat{\sigma}^2 = \frac{1}{T-1} \sum_{t=1}^{T} (r_t - \hat{\mu})^2
$$

- These summarize the data into a few representative values.
- They are deterministic once the dataset is fixed.


### Estimator
An **estimator** is a random variable:

- It depends on future (random) samples.
- It has a **sampling distribution**.
- Its mean and variance describe estimation uncertainty.

### Estimate
An **estimate** is the numerical value obtained from an estimator on a specific dataset.

Example:
- Estimator: sample mean formula
- Estimate: $\hat{\mu} = 0.0012$

We often assume a model for returns:
$$
r_t = \mu + \sigma z_t, \quad z_t \sim \mathcal{N}(0,1)
$$

- $\mu$: drift (expected return)
- $\sigma$: volatility

### Maximum Likelihood Estimation (MLE)

Under normality, the MLE estimators are:
$$
\hat{\mu} = \frac{1}{T} \sum_{t=1}^{T} r_t
$$
$$
\hat{\sigma}^2 = \frac{1}{T-1} \sum_{t=1}^{T} (r_t - \hat{\mu})^2
$$

- For large $T$, these estimators converge to the true parameters.
- These are the standard estimators used in practice.



## 3. Simple Returns vs Log Returns

### Definitions

- Simple return:
$$
R_t = \frac{P_t - P_{t-1}}{P_{t-1}}
$$

- Log return:
$$
r_t = \log\left(\frac{P_t}{P_{t-1}}\right)
$$

### Relationship

$$
R_t = e^{r_t} - 1
$$

### Why use log returns?

- They are additive over time:
$$
\log\left(\frac{P_T}{P_0}\right) = \sum_{t=1}^{T} r_t
$$
- They are approximately normally distributed in many models.
- They simplify theoretical modeling (e.g., GBM).






## 4. Mean and Variance of Simple Returns

If log returns are normal:
$$
r_t \sim \mathcal{N}(\mu, \sigma^2)
$$

Then:
$$
\mathbb{E}[R_t] = e^{\mu + \frac{1}{2}\sigma^2} - 1
$$
$$
\mathrm{Var}(R_t) = e^{2\mu} \left(e^{\sigma^2} - 1\right)
$$

- Mean and variance of simple returns are **nonlinear** functions of $\mu$ and $\sigma$.
- Therefore, simple returns introduce distortions.



## 5. Taylor Approximation (Small Returns)



For small $r_t$:
$$
R_t = e^{r_t} - 1 \approx r_t
$$

Then:
$$
\mathbb{E}[R_t] \approx \mu
$$
$$
\mathrm{Var}(R_t) \approx \sigma^2
$$

- This approximation works only when returns are small.
- For realistic volatility levels, differences can be significant.




## 6. Volatility Estimation



We model prices as:
$$
P_t = P_{t-1} e^{r_t}
$$
$$
r_t = \mu + \sigma z_t
$$

Key points:

- Volatility $\sigma$ is the key risk parameter.
- Drift $\mu$ does not directly affect variance, but must still be estimated.
- Log returns should be used for consistency with the model.



## 7. Alternative Volatility Estimators



Standard estimators use only close-to-close returns.

However, more efficient estimators use **OHLC data**:
- Open ($O_t$)
- High ($H_t$)
- Low ($L_t$)
- Close ($C_t$)

### Parkinson Estimator

Uses high-low range:
$$
\hat{\sigma}^2_{\text{Parkinson}} = \frac{1}{4T \log 2} \sum_{t=1}^{T} \left( \log\left(\frac{H_t}{L_t}\right) \right)^2
$$

- More efficient than close-to-close estimator.
- Assumes zero drift.

### Garman-Klass Estimator

Uses OHLC data:
$$
\hat{\sigma}^2_{\text{GK}} = \frac{1}{T} \sum_{t=1}^{T} \left[ 
\frac{1}{2} \left(\log\left(\frac{H_t}{L_t}\right)\right)^2 
- (2 \log 2 - 1) \left(\log\left(\frac{C_t}{O_t}\right)\right)^2
\right]
$$

- Incorporates more information → higher efficiency.
- Widely used in practice.

## 8. Hypothesis testing


Hypothesis testing is a formal statistical procedure used to determine if sample data provides sufficient evidence to reject a specific claim about a population. It evaluates two mutually exclusive statements, the null hypothesis (no effect) and the alternative hypothesis (an effect exists), to make data-driven decisions while managing the risk of errors.

- Null Hypothesis ($H_0$): Assumes there is no effect, no difference, or no relationship, often representing the status quo.
- Alternative Hypothesis ($H_a$): The statement that contradicts and is what the researcher is trying to prove.
- Significance Level ($\alpha$): The probability of rejecting the null hypothesis when it is true (Type I error). Common levels are 0.05 or 0.01.
- P-value: The probability of obtaining results at least as extreme as the observed data, assuming $H_0$ is true. If $P \leq \alpha, H_0$ is rejected.

### The 4-Step Process
1. State the Hypotheses: Define $H_0$ and $H_a$
2. Formulate an Analysis Plan: Determine the significance level ($\alpha$) and choose the appropriate test (Z-test, t-test, ecc...).
3. Analyze Sample Data: Calculate the test statistic and corresponding p-value.
4. Interpret Results: Compare the p-value to $\alpha$. If $p \leq \alpha$, reject $H_0$, otherwise fail to reject H_0.

### Types of Errors
- Type I error: rejecting $H_0$ when it is actually true (false positive).
- Type II error: failing to reject $H_0$ when it is actually false (false negative)


# Alternnatives to the random walk

## Autocorrelation

Analyzing market data for stock prices in the style of Lo and McKinley doing a variance ratio test, the presence of serial correlation in returns indicates thate the random walk model is not a good fit for the data. Rejecting the random walk means that the returns are not independent. If they were independent, the variance would grow linearly with time. If we observe departures from this behaviour we reject the random walk and we are given a clue of what we might want to use to replace it.

If the return are correlated across time, there will be non-vanishing cross-terms in the variance of the q-sum.

Let's define autocorrelation and autocovariance coefficients, which measure the correlation and covariance of a time series with **itself**, shifted in time by k steps. For a covariance-stationary (dependence between values only depends on how far apart they are, not when they occur) process, let:

$$\gamma_k = \text{Cov}(r_t, r_{t-k})$$
$$\gamma_0 = \text{Var}(r_t)$$
$$\rho_k = \text{Corr}(r_t, r_{t-k} = \frac{\gamma_k}{\gamma_0}$$

because the cross-terms are non-vanishing, the variance of the two-period returns is no longer double the 1-period:

$$\text{Var}(r_t^{(2)}) = \text{Var}(r_t + r_{t-1}) = 2\text{Var}(r_t) + 2\text{Cov}(r_t, r_{t-1}) = 2\text{Var}(r_t)(1 + \rho_1)$$

### Autocovariance estimation

If we want to estimate the autocorrelation at lag k using T pairs, then we need T+k observations:
$$\hat{\gamma_k} = \frac{1}{T-1} \sum^{t = k+1}_{k + T} (r_t - \hat{\mu})(r_{t-k} - \hat{\mu'})$$

where $\hat{\mu}$ is the estimate of the mean of the current sample : $\hat{\mu} = \frac{1}{T} \sum_{t=1}^{T} r_t$
and $\hat{\mu'}$ is the estimate of the mean of the k-shifted sample : $\hat{\mu}' = \frac{1}{T} \sum_{t=1}^{T} r_{t-k}$

For $T \rightarrow \infty$ then $\hat{\mu} \rightarrow \hat{\mu'}$ and:
$$\lim_{T \to \infty} \hat{\gamma}_k = \gamma_k$$
$$\sqrt{T}\,(\hat{\gamma}_k - \gamma_k) \sim \mathcal{N}(0, \sigma_\gamma^2)$$
$$\sqrt{T}\,(\hat{\rho}_k - \rho_k) \sim \mathcal{N}(0, 1)$$

Serial correlation in a time series will cause the cariance ratio to depart from unity:

$$\text{Var}(r_t + r_{t-1}) = 2\,\text{Var}(r_t) + 2\,\text{Cov}(r_t, r_{t-1})$$
$$\frac{\text{Var}(r_t^{(2)})}{2\,\text{Var}(r_t)} = 1 + \rho_1
\begin{cases}
= 1 & \text{random walk} \\
> 1 & \text{positive serial correlation} \\
< 1 & \text{negative serial correlation}
\end{cases}$$

We can generalize to higher (finite) correlation structure:

$$\text{VR}(q) = \frac{\text{Var}(r_t^{(q)})}{q\,\text{Var}(r_t)} 
= 1 + 2\sum_{k=1}^{q-1} \left(1 - \frac{k}{q}\right)\rho_k$$
