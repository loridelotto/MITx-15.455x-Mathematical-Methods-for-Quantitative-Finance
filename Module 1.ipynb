# 1 Random Variables



**Random variables** are functions that assign a probability to events in the sample space. They can be discrete, continuous, or a mix of both.

# 2 Probability Distributions


Continuous random variables: we ask for the probability to lie within a given range:
$$ Prob(a<X<b) = \int_{a}^{b} p(x) dx$$
where $p(x)$ is the **Probability Distribution** and needs to satisfy:
1) $p(x) \geq 0$
2) $\sum_{k}^{} p(x_k) = 1$ , $x_k$ discrete
3) $ \int_{-\infty}^{\infty} p(x) dx = 1 $ , x continuous
We also have the **Cumulative Distribution Function** that gives the probability that X is less than or equal to a given value:
$$F(x) = Prob(X<x) = \int_{-\infty}^{x} p(x') dx' $$
It contains nearly the same information as the probability density, since 
$$p(x) = \frac{dF(x)}{dx}$$
The CDF is often easier to approximate from empirical data and it is useful since 
$$Prob(a< X < b) = F(b) - F(a)$$
To change the variable from x to y(x) the density in terms of the new variable is given by 
$$p(x)dx = g(y) dy$$
which preserves the normalization condition (assuming y is an increasing function of x). If not, an absolute value is needed for positivity $g(y) = \frac{p(x)}{|dy/dx|}$

# 3 Expectation and moments

The **Expected value** E is the theoretical long-term average outcome of a random variable, calculated by multiplying each possible result by its probability and summing those values:
- For the discrete case: $$E[f(X)] = \sum_{k = 1}^{n} f(x_k)p(x_k)$$
- For the continuous case: $$E[f] = \int_{-\infty}^{\infty} f(x)p(x) dx$$
The **mean** of the distribution is simply the expectation of the random variable itself (if the sample space is infinite, the mean is not guaranteed to exist since the integral or the sum might not converge).

The **Moments** of a distribution are the expectation of powers of the random variable itself. As we will see later, if all the moments exist and are known, they can be used to get the expectation of other functions.

$$\mu_l = E[X^l] = \begin{cases}
\sum_{k}x^l_k p(x_k) \\
\int x^l p(x)dx
\end{cases}$$

The expectation operator is linear:
- $E[cf(X)] = cE[f(X)]$
- $E[f(X) + g(X)] = E[f(X)] + E[g(X)]$

In particular, the *second moment* is very interesting as it represents the **Variance**:
$$\sigma^2 = Var(X) = E[(X-\mu)^2] = E[X^2] - E[X]^2$$

The **Standard Deviation** is the square root of the variance and has the same units as the random variable.

**HIGHER MOMENTS**
- Variance: dispersion measure based on second moment $$\sigma^2 = \int (x-\mu)^2 p(x)dx$$
- Skewness: asymmetry parameter based on 3rd moments, dimensionless: $$s = \frac{E[(X-\mu)^3]}{\sigma^3}$$ 
    -  s < 0 long tail on the left
    -  s = 0 symmetrical distribution
    - s > 0 long tail on the right
    
- Kurtosis: measure of tail weight, 0 for Gaussian: $$k = \frac{E[(X-\mu)^4]}{\sigma^4} - 3$$

# 4 Covariance and correlation

For any two random variables, not necessarily independent or identically distributed, their **covariance** is defined as:
$$\text{Cov}(X,Y) = E[(X-\mu_x)(Y-\mu_y)] = E[XY] - \mu_x \mu_y$$
The **correlation** is proportional to the covariance: 
$$ \rho(X,Y) = \text{Corr}(X,Y) = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}} = E\left[ \left( \frac{X-\mu_x}{\sigma_x} \right) \left( \frac{Y-\mu_y}{\sigma_y} \right) \right] $$
with $$ -1 \leq \rho(X,Y) \leq +1$$
If the variables are independent, then the covariance will be 0. However the converse is not true: vanishing covariance does not imply independence.   
The correlation is only sensitive to linear dependence


# 5 Common distributions

## Uniform distribution

- Standard form: 
$$p(x) = \begin{cases} 1, & x \in [0, 1] \\ 0, & \text{otherwise} \end{cases}$$


$$\text{Prob}(a < X < b) = \int_{a}^{b} p(x)dx = (b - a)$$


$$F(x) = \int_{-\infty}^{x} p(x')dx' = x$$

- Moments:
$$\mu = \int_{-\infty}^{\infty} x p(x)dx = \int_{0}^{1} xdx = \frac{1}{2}$$

$$\mu_{\ell} = \int_{0}^{1} x^{\ell}dx = \frac{1}{\ell + 1}$$

$$\sigma^2 = \int_{0}^{1} \left(x - \frac{1}{2}\right)^2 dx = \frac{1}{12}$$

## Binomial distribution


Provides a model for any observation or experiment that has two possible outcomes "success" or "failure"
$$f(k; n, p) = \binom{n}{k} p^k q^{n-k}$$
where $q = 1-p$ is the probability of "failure"

$$\binom{n}{k} = \frac{n!}{k!(n-k)!} = \binom{n}{n-k}$$
gives the probability of k successes that have probability p of happening, in n tries.
It can be proven that $$\mu = E[X] = np$$


## Gaussian distribution

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{{-(x-\mu)^2}/{2\sigma^2}}$$

given $z = \frac{x-\mu}{\sigma}$ it is possible to rescale the Gaussian to: $$p(z) = \frac{1}{\sqrt{2\pi}}e^{-z^2/2}$$
$\mu$ represents the center of the distribution, while $\sigma$ represents the width.

HANDY FORMULAS:

- Basic Gaussian integral: $I(a) = \int_{-\infty}^{\infty} e^{-ax^2} dx = \sqrt{\frac{\pi}{a}}$
- Second moment through the derivative: $-\frac{dI(a)}{da} = \int_{-\infty}^{\infty} x^2 e^{-ax^2} dx = \frac{1}{2} \sqrt{\frac{\pi}{a^3}}$

Probability in a range computed from the cumulative distribution function:
$$F(z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{z} e^{-z'^2/2} dz' = \Phi\left(\frac{x - \mu}{\sigma}\right)$$
$$F(b) - F(a) = \text{Prob}(a < Z < b)$$

## Lognormal distribution


If we change variables $X = \log Y, \quad x = \log y$ we obtain the lognormal distribution function: 
$$g(y) = \frac{p(x)}{|dy/dx|} = \frac{1}{y\sqrt{2\pi\sigma^2}}e^{-(\log y - \mu)^2 / (2\sigma^2)}$$

## Poisson distribution


$$p(k;\lambda) = \frac{e^{-\lambda} \lambda^k}{k!}$$
Describes the probability to find exaclty k events during interval of time t:

$$p(k;\lambda t) = \frac{e^{-\lambda t}(\lambda t)^k}{k!}$$
where k is an integer and $\lambda$ is real.

**Mean**: $\mu = \lambda$

**Variance**: $\sigma^2 = \lambda$

The Poisson distribution can be used to approximate the Binomial distribution for $n \rightarrow \infty$ and $\lambda = np$

# 6 Sums of Random Variables


If we have two or more random variables, we can sum them to obtain new random variables.
When there are several random variables, it is often usegul to ask about the distribution of their sum:
$$S = X_1 + X_2 + \cdots + X_n$$
In simple cases, the full probability distribution can be derived. But in more complex cases, we only try to compute some of the moments of the full distribution, such as the *mean* and *variance*. This is easy to do thanks to the linearity of the expectation operator:
$$E[S] = E[X_1] + E[X_2] + \cdots + E[X_n]$$

Due to the **Central Limit Theorem** this is often enough. If there are enough variables, and if the individual distributions in the sum are well-behaved, then the sum will be an (appowimately) Gaussian variable, in which case the mean and variance describe it completely.



### Example: portfolio returns

Let a portfolio consist of N assets and let:
- $\omega_i$ = portfolio weight in asset $i$,
- $R_i$ be the return on asset &i&,
- $\mu_i = E[R_i], \sigma^2_i = \text{Var}(R_i)$

the return of the portfolio will be: $$R_p = \sum_{i = 1}^{N}\omega_i R_i = \omega_1 R_1 + \omega2 R_2 + \cdots + \omega_N R_N$$ 

while the expected value will be: $$\mu_p = E[R_p] = \sum_{i = 1}^{N}\omega_i E[R_i] = \sum_{i = 1}^{N}\omega_i \mu_i$$

We are not making assumptions about the individual return distributions except for the existence of the mean and variance.

The portfolio risk, measured by its standard deviation, does involve the covariance of pairs of asset returns.

$$\sigma_p^2 = \text{Var}(R_p) = \text{E}[(R_p - \mu_p)^2] = \text{E} \left[ \left( \sum_{i=1}^{N} w_i(R_i - \mu_i) \right)^2 \right]$$ $$= \sum_{i=1}^{N} w_i^2 \text{E}[(R_i - \mu_i)^2] + 2 \sum_{i < j}^{N} w_i w_j \text{E}[(R_i - \mu_i)(R_j - \mu_j)]$$ $$= \sum_{i=1}^{N} w_i^2 \text{Var}(R_i) + 2 \sum_{i < j}^{N} w_i w_j \text{Cov}(R_i, R_j) = \sum_{i=1}^{N} w_i^2 \sigma_i^2 + 2 \sum_{i < j}^{N} w_i w_j \sigma_i \sigma_j \rho_{ij}$$

- Correlation = 0, variance equals sum of squares $$\rho_{ij} = 0, \quad \sigma^2_p = \sum(\omega_i \sigma_i)^2$$
- Correlation = 1, variance equals square of sum $$\rho_{ij} = 1, \quad \sigma^2_p = \left(\sum\omega_i \sigma_i \right)^2$$
- Correlation = 0, all weights and variances identical $$\rho_{ij} = 0,\quad \omega_i = \frac{1}{N},\quad \sigma_i = \sigma_0 \ \Rightarrow \ \sigma_p^2 = \frac{\sigma_0^2}{N}$$

## Two Random variables


Two, however, is not a large number. 
Let $S = X_1 + X_2$ be the sum of two independent random variables with density functions $p_1(x_1)$ and $p_2(x_2)$. Then

$$F(x) = \text{Prob}(X_1 + X_2 < x) = \iint_A p_1(x_1)p_2(x_2) dx_1 dx_2$$
$$= \int_{x_1=-\infty}^{x_1=\infty} \int_{x_2=-\infty}^{x_2=x-x_1} p_1(x_1)p_2(x_2) dx_1 dx_2$$

So

$$p(x) = \frac{dF(x)}{dx} = \int_{-\infty}^{\infty} p_1(x_1)p_2(x - x_1) dx_1,$$

which defines the *convolution* of $p_1$ and $p_2$. In the case of the uniform distribution

$$p(x) = \begin{cases} 
x, & x \in [0, 1] \\ 
2 - x, & x \in [1, 2] \\ 
0, & \text{otherwise} 
\end{cases}$$

## Binomial Distribution

We can view the number of "successes" as a sum of random variables, one for each of the Bernoulli trials that has probability p, $$\text{Let}\ S = X_1 + X_2 + \cdots + X_n$$.
$$\text{Then}\ \ E[S] = E[X_1] + E[X_2] + \cdots + E[X_n] = np$$

This takes account of the fact that X's are **independent** and **indentically distributed**.

This approach makes the variance especially easy to compute:

\begin{equation*}
\begin{aligned}
\sigma^2 &= \text{E}[(S - \mu)^2] \\
&= \text{E}[(X_1 + X_2 + \dots + X_n)^2] - \mu^2 \\
&= n\text{E}[X_1^2] + n(n-1)\text{E}[X_1 X_2] - \mu^2 \\
&= np + n(n-1)p^2 - (np)^2, \quad \text{since } \text{E}[X_i X_j] = \begin{cases} p, & \text{if } i = j \\ p^2, & \text{if } i \neq j \end{cases} \\
&= np(1 - p) = npq,
\end{aligned}
\end{equation*}

$$ \sigma = \sqrt{npq} $$

# 7 The Characteristic Function



The **characteristic function** of a random variable $X$ is a complete description of its probability distribution. It is particularly useful because its derivatives at the origin generate all the moments of the distribution.


The characteristic function, denoted by $\tilde{f}(t)$, is defined as the expectation of $e^{itX}$:

$$\tilde{f}(t) \equiv \text{E}[e^{itX}]$$

The relationship between the $\ell$-th moment of $X$ and the derivatives of the characteristic function is given by:

$$\text{E}[X^\ell] = (-i)^\ell \frac{\text{d}^\ell}{\text{d}t^\ell} \tilde{f}(t) \bigg|_{t=0}$$

### Derivation of the Moment Relationship



To understand how the second relationship is derived, we use the property of expectation and differentiation:

1.  **Definition**: Start with the definition $\tilde{f}(t) = \text{E}[e^{itX}]$.
2.  **Differentiation**: Differentiate both sides $\ell$ times with respect to $t$. Assuming we can differentiate under the expectation sign:
    $$\frac{\text{d}^\ell}{\text{d}t^\ell} \tilde{f}(t) = \text{E}\left[\frac{\text{d}^\ell}{\text{d}t^\ell} e^{itX}\right]$$
3.  **Chain Rule**: Each derivative of $e^{itX}$ with respect to $t$ brings down a factor of $(iX)$:
    $$\frac{\text{d}^\ell}{\text{d}t^\ell} \tilde{f}(t) = \text{E}[(iX)^\ell e^{itX}] = i^\ell \text{E}[X^\ell e^{itX}]$$
4.  **Evaluation at $t=0$**: Set $t=0$. Since $e^0 = 1$:
    $$\frac{\text{d}^\ell}{\text{d}t^\ell} \tilde{f}(t) \bigg|_{t=0} = i^\ell \text{E}[X^\ell]$$
5.  **Isolation of the Moment**: Divide by $i^\ell$. Since $\frac{1}{i^\ell} = (-i)^\ell$, we obtain:
    $$\text{E}[X^\ell] = (-i)^\ell \frac{\text{d}^\ell}{\text{d}t^\ell} \tilde{f}(t) \bigg|_{t=0}$$


### Example: The Binomial Distribution



For a binomial distribution $B(n, p)$, where $q = 1-p$, the characteristic function is derived as follows:

$$\tilde{f}(t) = \text{E}[e^{itX}] = \sum_{k=0}^{n} e^{itk} \text{Prob}(X = k)$$
$$= \sum_{k=0}^{n} e^{itk} \binom{n}{k} p^k q^{n-k}$$
$$= \sum_{k=0}^{n} \binom{n}{k} (pe^{it})^k q^{n-k}$$
Using the Binomial Theorem, this simplifies to a closed form:
$$\tilde{f}(t) = (pe^{it} + q)^n$$

## Continuous distributions: Fourier transform of density


The characteristic function is just the Fourier Transform of the probability density:

$$
\tilde{p}(t) \equiv \int_{-\infty}^{\infty} e^{itx} p(x)\,dx
= \sum_{\ell=0}^{\infty} \frac{(it)^\ell}{\ell!}\, \mathbb{E}[X^\ell]
$$

$$
\mu_\ell = \mathbb{E}[X^\ell] = (-i)^\ell \left. \frac{d^\ell}{dt^\ell} \tilde{p}(t) \right|_{t=0}
$$

**Especially useful for sums of random variables:**

$$
Y = X_1 + X_2
$$

- Density function of the sum = **convolution** of the individual functions:
$$
p(y) = \int p_1(x_1)\, p_2(y - x_1)\, dx_1
$$

- Fourier transform of a convolution = **product** of the Fourier transforms:
$$
\tilde{p}(t) = \tilde{p}_1(t)\, \tilde{p}_2(t)
$$

# 8 Central Limit Theorem

The Fourier transform of a Gaussian is also a Gaussian:
$$p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{(x-\mu)/{2 \sigma^2}}, \quad \quad \tilde{p}(t) = e^{-\sigma^2 t^2/2 +i \mu t}$$

The sum of N independent Gaussian random variables is itself a Gaussian RV with its own mean and variance:
$$
\tilde{p}(t) = \tilde{p}_1(t)\tilde{p}_2(t)\cdots \tilde{p}_N(t)
$$

$$
= \exp\left[-\frac{\sigma_1^2 t^2}{2} + i\mu_1 t \right]
\exp\left[-\frac{\sigma_2^2 t^2}{2} + i\mu_2 t \right] \cdots
$$

$$
= \exp\left[
-\frac{(\sigma_1^2 + \cdots + \sigma_N^2)t^2}{2}
+ i(\mu_1 + \cdots + \mu_N)t
\right]
$$

$$
= \exp\left[
-\frac{\hat{\sigma}^2 t^2}{2}
+ i\hat{\mu} t
\right]
$$

where

$$
\hat{\sigma}^2 = \sum \sigma_i^2, \qquad
\hat{\mu} = \sum \mu_i
$$

If the Gaussians are **identical** the moments can be computed as follows:
$$\mu \rightarrow \hat{\mu} = N \mu$$
$$\sigma \rightarrow \hat{\sigma} = \sqrt{N}\sigma$$

### Cumulant expansion

Since Fourier transforms multiply, their logarithms add**. This motivates the definition of the **cumulant generating function** as the logarithm of the characteristic function.

$$
\tilde{p}(t) \equiv \mathbb{E}[e^{itX}]
= \exp\left[
\sum_{\ell=0}^{\infty} \frac{(it)^\ell}{\ell!} C_\ell
\right]
$$

The coefficients $C_n$ are called **cumulants**, and are defined as derivatives of the log-characteristic function:

$$
C_n = (-i)^n \left. \frac{d^n}{dt^n} \log \tilde{p}(t) \right|_{t=0}
$$


Interpretation:

- The function $\tilde{p}(t)$ is the **characteristic function** (Fourier transform of the density).
- Taking the logarithm isolates the part that becomes **additive** for sums of independent variables.
- Therefore, **cumulants add** when independent random variables are summed, unlike raw moments.

Relation to moments:

Moments are obtained from derivatives of $\tilde{p}(t)$, while cumulants come from derivatives of $\log \tilde{p}(t)$.  
Expanding:

$$
\log \tilde{p}(t)
= \log \mathbb{E}[e^{itX}]
$$

and using a Taylor expansion around $t=0$ leads to the cumulant series.


First few cumulants:

$$
C_1 = \langle X \rangle
$$

$$
C_2 = \langle X^2 \rangle - \langle X \rangle^2
$$

$$
C_3 = \langle X^3 \rangle - 3\langle X \rangle \langle X^2 \rangle + 2\langle X \rangle^3
$$

$$
C_4 = \langle X^4 \rangle - 3\langle X^2 \rangle^2 - 4\langle X \rangle \langle X^3 \rangle + 12\langle X \rangle^2 \langle X^2 \rangle - 6\langle X \rangle^4
$$

Key properties:

- $C_1$ = mean  
- $C_2$ = variance  
- Higher cumulants measure deviations from Gaussianity:
  - $C_3$ → skewness  
  - $C_4$ → kurtosis  

- For a Gaussian distribution:
$$
C_n = 0 \quad \text{for all } n > 2
$$

 Why this works (intuition)

If $X = X_1 + \cdots + X_N$ with independent variables:

$$
\tilde{p}(t) = \prod_i \tilde{p}_i(t)
\quad \Rightarrow \quad
\log \tilde{p}(t) = \sum_i \log \tilde{p}_i(t)
$$

So cumulants (coefficients of the expansion of $\log \tilde{p}(t)$) are additive, making them especially useful for studying sums of random variables.

**Cumulants for sums of i.i.d. variables**

Consider the cumulants for a sum of $N$ independent, identically distributed random variables:
$$
S_N = X_1 + \cdots + X_N
$$

Since cumulants add for independent variables:
$$
\hat{C}_n = N C_n
$$

Normalized cumulants

Define the variance of the sum:
$$
\hat{\sigma}^2 = N \sigma^2
$$

Then the dimensionless, normalized cumulants scale as:
$$
\frac{\hat{C}_n}{\hat{\sigma}^n}
= \frac{1}{N^{\,\frac{n}{2}-1}} \left( \frac{C_n}{\sigma^n} \right)
$$

Central Limit behavior

As $N \to \infty$:
$$
\hat{C}_n \to 0 \quad \text{for } n > 2
$$

So the distribution of the sum approaches a Gaussian, since only:
- $C_1$ (mean)
- $C_2$ (variance)

remain nonzero.

Important remark

This result explains *why* convergence to a Gaussian happens, but: it does not describe the rate of convergence in different regions of the distribution
