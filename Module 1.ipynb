# 1 Random Variables



**Random variables** are functions that assign a probability to events in the sample space. They can be discrete, continuous, or a mix of both.

# 2 Probability Distributions


Continuous random variables: we ask for the probability to lie within a given range:
$$ Prob(a<X<b) = \int_{a}^{b} p(x) dx$$
where $p(x)$ is the **Probability Distribution** and needs to satisfy:
1) $p(x) \geq 0$
2) $\sum_{k}^{} p(x_k) = 1$ , $x_k$ discrete
3) $ \int_{-\infty}^{\infty} p(x) dx = 1 $ , x continuous
We also have the **Cumulative Distribution Function** that gives the probability that X is less than or equal to a given value:
$$F(x) = Prob(X<x) = \int_{-\infty}^{x} p(x') dx' $$
It contains nearly the same information as the probability density, since 
$$p(x) = \frac{dF(x)}{dx}$$
The CDF is often easier to approximate from empirical data and it is useful since 
$$Prob(a< X < b) = F(b) - F(a)$$
To change the variable from x to y(x) the density in terms of the new variable is given by 
$$p(x)dx = g(y) dy$$
which preserves the normalization condition (assuming y is an increasing function of x). If not, an absolute value is needed for positivity $g(y) = \frac{p(x)}{|dy/dx|}$

# 3 Expectation and moments

The **Expected value** E is the theoretical long-term average outcome of a random variable, calculated by multiplying each possible result by its probability and summing those values:
- For the discrete case: $$E[f(X)] = \sum_{k = 1}^{n} f(x_k)p(x_k)$$
- For the continuous case: $$E[f] = \int_{-\infty}^{\infty} f(x)p(x) dx$$
The **mean** of the distribution is simply the expectation of the random variable itself (if the sample space is infinite, the mean is not guaranteed to exist since the integral or the sum might not converge).

The **Moments** of a distribution are the expectation of powers of the random variable itself. As we will see later, if all the moments exist and are known, they can be used to get the expectation of other functions.

$$\mu_l = E[X^l] = \begin{cases}
\sum_{k}x^l_k p(x_k) \\
\int x^l p(x)dx
\end{cases}$$

The expectation operator is linear:
- $E[cf(X)] = cE[f(X)]$
- $E[f(X) + g(X)] = E[f(X)] + E[g(X)]$

In particular, the *second moment* is very interesting as it represents the **Variance**:
$$\sigma^2 = Var(X) = E[(X-\mu)^2] = E[X^2] - E[X]^2$$

The **Standard Deviation** is the square root of the variance and has the same units as the random variable.

**HIGHER MOMENTS**
- Variance: dispersion measure based on second moment $$\sigma^2 = \int (x-\mu)^2 p(x)dx$$
- Skewness: asymmetry parameter based on 3rd moments, dimensionless: $$s = \frac{E[(X-\mu)^3]}{\sigma^3}$$ 
    -  s < 0 long tail on the left
    -  s = 0 symmetrical distribution
    - s > 0 long tail on the right
    
- Kurtosis: measure of tail weight, 0 for Gaussian: $$k = \frac{E[(X-\mu)^4]}{\sigma^4} - 3$$

# 4 Covariance and correlation

For any two random variables, not necessarily independent or identically distributed, their **covariance** is defined as:
$$\text{Cov}(X,Y) = E[(X-\mu_x)(Y-\mu_y)] = E[XY] - \mu_x \mu_y$$
The **correlation** is proportional to the covariance: 
$$ \rho(X,Y) = \text{Corr}(X,Y) = \frac{\text{Cov}(X,Y)}{\sqrt{\text{Var}(X)\text{Var}(Y)}} = E\left[ \left( \frac{X-\mu_x}{\sigma_x} \right) \left( \frac{Y-\mu_y}{\sigma_y} \right) \right] $$
with $$ -1 \leq \rho(X,Y) \leq +1$$
If the variables are independent, then the covariance will be 0. However the converse is not true: vanishing covariance does not imply independence.   
The correlation is only sensitive to linear dependence


# 5 Common distributions

## Uniform distribution

- Standard form: 
$$p(x) = \begin{cases} 1, & x \in [0, 1] \\ 0, & \text{otherwise} \end{cases}$$


$$\text{Prob}(a < X < b) = \int_{a}^{b} p(x)dx = (b - a)$$


$$F(x) = \int_{-\infty}^{x} p(x')dx' = x$$

- Moments:
$$\mu = \int_{-\infty}^{\infty} x p(x)dx = \int_{0}^{1} xdx = \frac{1}{2}$$

$$\mu_{\ell} = \int_{0}^{1} x^{\ell}dx = \frac{1}{\ell + 1}$$

$$\sigma^2 = \int_{0}^{1} \left(x - \frac{1}{2}\right)^2 dx = \frac{1}{12}$$

## Binomial distribution


Provides a model for any observation or experiment that has two possible outcomes "success" or "failure"
$$f(k; n, p) = \binom{n}{k} p^k q^{n-k}$$
where $q = 1-p$ is the probability of "failure"

$$\binom{n}{k} = \frac{n!}{k!(n-k)!} = \binom{n}{n-k}$$
gives the probability of k successes that have probability p of happening, in n tries.
It can be proven that $$\mu = E[X] = np$$


## Gaussian distribution

$$p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{{-(x-\mu)^2}/{2\sigma^2}}$$

given $z = \frac{x-\mu}{\sigma}$ it is possible to rescale the Gaussian to: $$p(z) = \frac{1}{\sqrt{2\pi}}e^{-z^2/2}$$
$\mu$ represents the center of the distribution, while $\sigma$ represents the width.

HANDY FORMULAS:

- Basic Gaussian integral: $I(a) = \int_{-\infty}^{\infty} e^{-ax^2} dx = \sqrt{\frac{\pi}{a}}$
- Second moment through the derivative: $-\frac{dI(a)}{da} = \int_{-\infty}^{\infty} x^2 e^{-ax^2} dx = \frac{1}{2} \sqrt{\frac{\pi}{a^3}}$

Probability in a range computed from the cumulative distribution function:
$$F(z) = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{z} e^{-z'^2/2} dz' = \Phi\left(\frac{x - \mu}{\sigma}\right)$$
$$F(b) - F(a) = \text{Prob}(a < Z < b)$$

## Lognormal distribution


If we change variables $X = \log Y, \quad x = \log y$ we obtain the lognormal distribution function: 
$$g(y) = \frac{p(x)}{|dy/dx|} = \frac{1}{y\sqrt{2\pi\sigma^2}}e^{-(\log y - \mu)^2 / (2\sigma^2)}$$

## Poisson distribution


$$p(k;\lambda) = \frac{e^{-\lambda} \lambda^k}{k!}$$
Describes the probability to find exaclty k events during interval of time t:

$$p(k;\lambda t) = \frac{e^{-\lambda t}(\lambda t)^k}{k!}$$
where k is an integer and $\lambda$ is real.

**Mean**: $\mu = \lambda$

**Variance**: $\sigma^2 = \lambda$

The Poisson distribution can be used to approximate the Binomial distribution for $n \rightarrow \infty$ and $\lambda = np$
